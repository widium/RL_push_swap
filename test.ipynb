```python
STACK_SIZE = 3
NBR_STACK = 2
EPISODE = 20

env = Env(STACK_SIZE)
agent = DQNAgent(STACK_SIZE, NBR_STACK, len(env.action_space()))
```

```python
def plot_action(current_action : int):
    if current_action == 0:
        return "pb"
    elif current_action == 1:
        return "pa"
    elif current_action == 2:
        return "rb"
    elif current_action == 3:
        return "ra"
    elif current_action == 4:
        return "rrb"
    elif current_action == 5:
        return "rra"
    elif current_action == 6:
        return "sb"
    elif current_action == 7:
        return "sa"
    elif current_action == 8:
        return "rr"
    elif current_action == 9:
        return "rrr"
    elif current_action == 10:
        return "ss"    
```

```python
def predict(env : Env, agent : DQNAgent):
    
    list_actions = list()
    
    print_state(env)
    
    while (1):
        
        state = env.get_state()
        state = pre_processing_state(state)
        
        vector = agent.policy_model.predict(state)
        print_vector(vector)
        
        env.current_action = np.argmax(vector)
        action_name = plot_action(env.current_action)
        list_actions.append(action_name)
        env.take_actions()
        
        print_state(env)
        
        if (env.state() == 'done'):
            print_list_action(list_actions)
            break
    
```

```python
predict(env, agent)
```

```python
def take_actions(env : Env):
    
    if env.current_action == 0:
        env.moover.push_b(env.A, env.B)
        
    elif env.current_action == 1:
        env.moover.push_a(env.A, env.B)
        
    elif env.current_action == 2:
        env.moover.rotate_b(env.A, env.B)
        
    elif env.current_action == 3:
        env.moover.rotate_a(env.A, env.B)
        
    elif env.current_action == 4:
        env.moover.inverse_rotate_b(env.A, env.B)
        
    elif env.current_action == 5:
        env.moover.inverse_rotate_a(env.A, env.B)
        
    elif env.current_action == 6:
        env.moover.swap_b(env.A, env.B)
        
    elif env.current_action == 7:
        env.moover.swap_a(env.A, env.B)
        
    elif env.current_action == 8:
        env.moover.rotate(env.A, env.B)
        
    elif env.current_action == 9:
        env.moover.reverse(env.A, env.B)
        
    elif env.current_action == 10:
        env.moover.swap(env.A, env.B)
```
